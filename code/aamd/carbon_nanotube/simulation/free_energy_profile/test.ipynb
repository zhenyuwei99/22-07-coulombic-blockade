```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load(target_dir):
    res_list = [np.load(os.path.join(target_dir, i)) for i in os.listdir(target_dir) if 'npy' in i]
    return np.sum(np.stack(res_list), 0)
```

```python
target_dir = '/home/zhenyuwei/simulation_data/22-07-coulombic-blockade/code/aamd/carbon_nanotube/simulation/free_energy_profile/out/metad/no-wall-charge/POT-1.00e-01molPerL/pore-r0-8.125A-w0-50.000A-l0-50.000A-ls-25.000A-pot/04-sample-metad'
res = load(target_dir)
print(res.shape)
r = 51.577 * np.sqrt(2)
r = np.linspace(0, r/2 + 1, res.shape[0], endpoint=True)
z = 91.45
z = np.linspace(0, z/2 + 1, res.shape[1], endpoint=True)
r, z = np.meshgrid(r, z, indexing='ij')
fig, ax = plt.subplots(1, 1)
c = ax.contourf(r, z, res, 200)
fig.colorbar(c)
# c = ax.plot(z[0, :], res[0:30, :].mean(0))
```
